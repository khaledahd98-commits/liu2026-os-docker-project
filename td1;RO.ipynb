```python
#exo 1
import pulp

# 1. Création du problème (maximisation)
prob = pulp.LpProblem("Maximisation_Profit_Tapis", pulp.LpMaximize)

# 2. Variables de décision (continues, >= 0)
x1 = pulp.LpVariable("x1_Standard", lowBound=0)
x2 = pulp.LpVariable("x2_Luxe", lowBound=0)

# 3. Fonction objectif
prob += 300 * x1 + 400 * x2, "Profit_total"

# 4. Contraintes
prob += 2 * x1 + 1 * x2 <= 8, "Tissage"
prob += 1 * x1 + 2 * x2 <= 7, "Finition"

# 5. Résolution
prob.solve()

# 6. Résultats
print("Statut :", pulp.LpStatus[prob.status])
print(f"x1 (Standard) = {x1.varValue}")
print(f"x2 (Luxe)     = {x2.varValue}")
print(f"Profit optimal Z = {pulp.value(prob.objective)} MRU")

# Vérification des contraintes saturées
print("\nUtilisation des ressources :")
print(f"Tissage utilisé  : {2*x1.varValue + x2.varValue} % 8 h")
print(f"Finition utilisée: {x1.varValue + 2*x2.varValue} % 7 h")
```

```python
#exo 2
import pulp

# 1. Création du problème (minimisation)
prob = pulp.LpProblem("Minimisation_Cout_Approvisionnement", pulp.LpMinimize)

# 2. Variables de décision (continues, >= 0)
x1 = pulp.LpVariable("x1_ProduitA", lowBound=0)
x2 = pulp.LpVariable("x2_ProduitB", lowBound=0)

# 3. Fonction objectif
prob += 2 * x1 + 3 * x2, "Cout_total"

# 4. Contraintes
prob += x1 + x2 >= 6, "Contrainte_1"
prob += 2 * x1 + x2 >= 8, "Contrainte_2"

# 5. Résolution
prob.solve()

# 6. Résultats
print("Statut :", pulp.LpStatus[prob.status])
print(f"x1 (Produit A) = {x1.varValue}")
print(f"x2 (Produit B) = {x2.varValue}")
print(f"Cout minimal W = {pulp.value(prob.objective)}")

# Vérification des contraintes
print("\nVérification des contraintes :")
print(f"x1 + x2   = {x1.varValue + x2.varValue}  (>= 6)")
print(f"2x1 + x2  = {2*x1.varValue + x2.varValue}  (>= 8)")
```

```python
import pulp
print("=" * 55)
print("EXERCICE 4 : Domaine non borne")
print("=" * 55)

prob4 = pulp.LpProblem("Domaine_Non_Borne", pulp.LpMaximize)
y1 = pulp.LpVariable("x1", lowBound=0)
y2 = pulp.LpVariable("x2", lowBound=0)

prob4 += 4 * y1 + 2 * y2, "Z"
prob4 += y1 - y2 <= 2, "C1"
prob4 += 2 * y1 >= 4, "C2"

prob4.solve()
print("Statut :", pulp.LpStatus[prob4.status])
print("-> Le statut attendu est 'Unbounded' : Z croit indefiniment quand x2 augmente.\n")
```

```python
print("=" * 55)
print("EXERCICE 5 : Sensibilite sur C1")
print("=" * 55)

def resoudre(c1_val):
    prob5 = pulp.LpProblem("Sensibilite_C1", pulp.LpMaximize)
    a = pulp.LpVariable("x1", lowBound=0)
    b = pulp.LpVariable("x2", lowBound=0)
    prob5 += c1_val * a + 3 * b, "Z"
    prob5 += 2 * a + b <= 10, "C1"
    prob5 += a + 2 * b <= 8, "C2"
    prob5.solve()
    return a.varValue, b.varValue, pulp.value(prob5.objective)

# Test pour differentes valeurs de C1 : en-dessous, dans, et au-dessus de [1.5, 6]
for c1_test in [1.0, 1.5, 3.0, 6.0, 7.0]:
    x1_val, x2_val, z_val = resoudre(c1_test)
    print(f"C1 = {c1_test:>4} -> x1 = {x1_val:>4}, x2 = {x2_val:>4}, Z = {z_val}")

print("\n-> Le sommet (4, 2) reste optimal pour C1 dans [1.5, 6].")
print("   En dehors de cet intervalle, le sommet optimal change ((5,0) ou (0,4)).")
```

```python
#exo 6
import numpy as np
from scipy.optimize import linprog

# 1. Définition des indices
usines = [0, 1]       # U1, U2
entrepots = [0, 1, 2] # E1, E2, E3

# 2. Données du problème
couts = np.array([
    [5, 2, 4],  # U1 -> E1, E2, E3
    [3, 4, 6]   # U2 -> E1, E2, E3
])

capacites = [500, 700]
demandes = [300, 400, 350]

# Flatten de la matrice des coûts pour scipy (vecteur c)
c = couts.flatten()

# 3. Construction des contraintes de capacité (Offre : <=)
A_ub = []
b_ub = []

for i in usines:
    ligne = np.zeros((len(usines), len(entrepots)))
    ligne[i, :] = 1
    A_ub.append(ligne.flatten())
    b_ub.append(capacites[i])

# 4. Construction des contraintes de demande (Demande : >=, donc <= en inversant les signes)
A_lesseq = []
b_lesseq = []

for j in entrepots:
    ligne = np.zeros((len(usines), len(entrepots)))
    ligne[:, j] = -1
    A_lesseq.append(ligne.flatten())
    b_lesseq.append(-demandes[j])

# Fusion des contraintes de type <=
A_total = np.vstack([A_ub, A_lesseq])
b_total = b_ub + b_lesseq

# 5. Bornes des variables (x_ij >= 0)
bornes = [(0, None) for _ in range(len(c))]

# 6. Résolution
res = linprog(c, A_ub=A_total, b_ub=b_total, method='highs')

# --- Affichage des résultats ---
if res.success:
    print(f"Statut : Optimisation réussie")
    print(f"Coût total minimum : {res.fun} €\n")

    # Redimensionnement des variables sous forme de matrice pour la lecture
    flux = res.x.reshape((len(usines), len(entrepots)))

    for i in usines:
        for j in entrepots:
            if flux[i, j] > 0:
                print(f"Flux Usine U{i+1} -> Entrepôt E{j+1} : {flux[i, j]:.0f} unités")
else:
    print("Le problème n'a pas pu être résolu :", res.message)
```

```python
#exo 7
import numpy as np
from scipy.optimize import linprog

# 1. Données stockées sous forme de dictionnaire (extensible à N composants)
composants = {
    'A': {'cout': 20, 'solvant': 0.10, 'pigment': 0.60},
    'B': {'cout': 30, 'solvant': 0.20, 'pigment': 0.40},
    'C': {'cout': 15, 'solvant': 0.40, 'pigment': 0.20}
}

# Paramètres globaux du problème
quantite_totale = 1000
solvant_min_pct = 0.25
pigment_max_pct = 0.40

# Liste des clés pour conserver l'ordre
cles = list(composants.keys())

# 2. Vecteur des coûts (Fonction objectif)
c = [composants[k]['cout'] for k in cles]

# 3. Contraintes d'inégalités (A_ub * x <= b_ub)
A_ub = []
b_ub = []

# Contrainte Solvant : >= 25% du total -> On inverse le signe pour avoir du <=
# (ex: -0.10*xA - 0.20*xB - 0.40*xC <= -250)
ligne_solvant = [-composants[k]['solvant'] for k in cles]
A_ub.append(ligne_solvant)
b_ub.append(-(solvant_min_pct * quantite_totale))

# Contrainte Pigment : <= 40% du total
ligne_pigment = [composants[k]['pigment'] for k in cles]
A_ub.append(ligne_pigment)
b_ub.append(pigment_max_pct * quantite_totale)

# 4. Contraintes d'égalités (A_eq * x == b_eq)
# Mélange total = 1000 kg
A_eq = [[1 for _ in cles]]
b_eq = [quantite_totale]

# 5. Bornes des variables (x >= 0)
bornes = [(0, None) for _ in cles]

# 6. Résolution du programme linéaire
res = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=bornes, method='highs')

# --- Affichage des résultats ---
if res.success:
    print("Optimisation réussie !")
    print(f"Coût total minimal : {res.fun:,.2f} €")
    print("\nComposition optimale du mélange :")
    for idx, k in enumerate(cles):
        print(f" - Composant {k} : {res.x[idx]:.2f} kg")
else:
    print("Le problème n'a pas pu être résolu :", res.message)
```
